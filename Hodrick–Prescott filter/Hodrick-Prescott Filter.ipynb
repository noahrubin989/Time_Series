---
The Hodrick Prescott Filter is a data smoothing technique that aims to extract the trend $\tau_t$ and cyclical component $c_t$ from a time series $y_t$ such that $y_t = \tau_t + c_t$. It aims to remove short term fluctuations associated with the time series in an attempt to reveal the general long term trends, which can then assist with economic forecasting. 

---

The statsmodels implementation is based off an academic paper ["On Adjusting the Hodrick-Prescott Filter for the Frequency of Observations"](https://home.uchicago.edu/~huhlig/papers/uhlig.ravn.res.2002.pdf) by Morten O. Ravn and Harald Uhlig. The implementation aims to find the componenets of the $\vec{\hat \tau} = \begin{pmatrix}\hat{\tau}_1\\\hat{\tau}_2\\\hat{\tau}_3\\\vdots\\\hat{\tau}_T\end{pmatrix}$ column vector that minimises the quadratic loss function 

$$\min_{\\{ \tau_{t}\\} }\sum_{t=1}^{T}c_{t}^{2}+\lambda\sum_{t=1}^{T}\left[\left(\tau_{t+1}-\tau_{t}\right)-\left(\tau_{t}-\tau_{t-1}\right)\right]^{2}$$

whereby $c_t^2 = (y_t - \tau_t)^2$ represents the squared deviations of the trend component from the actual values which penalises the cyclical component $c_t$. The $\lambda$ constant is a positive scalar multiple which penalises variability in the trend component series to encourage data smoothing. The larger the value of $\lambda$, the smoother is the solution series. As $\lambda \to \infty$, the trend acceleration approaches towards 0, resulting a line representing a least squares fit to the data. Based off suggestions made by Ravn and Uhlig in their paper (linked above and below), statsmodels suggests to use a $\lambda$ value of 6.25 for annual data, 1600 for quarterly data and 129600 for monthly data. For other time periods, using the HP filter might not be the best option.

---

For a (finite) sample of size $T$ our $\vec{\hat\tau}$ vector can be obtained through computing 

$$\vec{\hat\tau} = (I_T + \lambda K^TK)^{-1}\vec{y}$$

whereby statsmodels defines $I_T$ as an identity matrix that has dimension $T \times T$, where $T$ is the number of rows in the dataset. The matrix $K$ is a $(T - 2) \times T$ matrix that has the following properties:

* $K_{ij} = -2$ when $i = j + 1$
* $K_{ij} = 1$ in cases where $i = j$ or $i = j + 2$
* $K_{ij} = 0$ everywhere else

where $K_{ij}$ represents row $i$, column $j$ in our matrix $K$

Computing $K^T K$ ensures a $T \times T$ matrix, which means that we are able to add this matrix to $I_T$. The fact that this $K^TK$ matrix is square ensures that $(I + K^TK)$ is also square, hence we can find an inverse (it exists). The presence of many zeroes in the $K$ matrix makes this a [sparse matrix](https://machinelearningmastery.com/sparse-matrices-for-machine-learning/), which can be space inefficient for larger sample sizes. Since $\vec{\hat\tau} = (I_T + \lambda K^TK)^{-1}\vec{y} $  is a process of multiplying a matrix and a vector toghether, by matrix multiplication laws, each component of the trend estimate vector can be computed through a weighted sum whereby 

$$\hat{\tau}_i = \sum_{j=1}^T p_{ij}y_j$$

in which $p_{ij}$ represents row $i$, column $j$ in the $(I_T + \lambda K^TK)^{-1}$ matrix. Since $\hat{c}_i = (y_i - \hat{\tau}_i)$ is computed based off our trend estimates, the vector of estimated cyclical components of the time series can also be viewed as a weighted sum of the individual values $y_i$ in our time series.
